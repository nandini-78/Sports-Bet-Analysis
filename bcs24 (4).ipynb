# 11. Can we identify any trends or patterns in the data that can help us optimize the betting experience for users?

```python
df1=bet_event_sport[bet_event_sport['outcome']==0].groupby(bet_event_sport.start_date.dt.year).agg({"outcome":"count"})
```

```python
df1
```

```python
df2=bet_event_sport[bet_event_sport['outcome']==1].groupby(bet_event_sport.start_date.dt.year).agg({"outcome":"count"})
```

```python
df2
```

```python
fig, ax = plt.subplots(figsize=(15,6))
sns.lineplot(data=df1, x="start_date", y="outcome")
plt.xlabel("Sports", size=15)
plt.ylabel("counts of bet lost", size=15)
plt.title("Bet lost over the year", size=18)
```

```python
fig, ax = plt.subplots(figsize=(15,6))
sns.lineplot(data=df2, x="start_date", y="outcome")
plt.xlabel("Sports", size=15)
plt.ylabel("counts of bet lost", size=15)
plt.title("Bet Won over the year", size=18)
```

```python
df3=pd.concat([df1, df2], axis=1).reindex(df1.index)
```

```python
df3
```

```python
df3.rename(columns={'start_date':'Year','outcome':'Bet lost', 'outcome':'Bet Won'}, inplace = True)
```

```python
df3
```

```python
df3.shape
```

```python
plt.figure()
df3.plot(figsize=(10,6))
plt.legend(["bet lost","bet won"])
plt.show()
```
